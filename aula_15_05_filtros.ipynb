```
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import requests
from io import BytesIO
```

```
response = requests.get('https://www.placecage.com/640/480')
img = Image.open(BytesIO(response.content)).convert()
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
# transformar a imagem em cinza
gray_arr = np.mean(np.array(img), axis=2)
img_gray = Image.fromarray(gray_arr)
plt.imshow(img_gray)
plt.axis('off')
plt.show()
```

```
# aplicar o filtro de suavização baseado na média 3x3, 5x5, (2k+1 x 2k+1)
# e plotar as imagens original e suavizada, lado a lado.
img_gray_copy = np.copy(img_gray)
filtro3x3 = np.ones((3,3))/9
filtro5x5 = np.ones((5,5))/25
# 1, 1, 1
# 1, 1, 1
# 1, 1, 1
# filtro3x3
img_filtered_arr = np.convolve(filtro5x5.flatten(), img_gray_copy.flatten(), 'same')
img_filtered_arr = img_filtered_arr.reshape((480, 640))
img_filtered = Image.fromarray(img_filtered_arr)
plt.imshow(img_filtered)
plt.axis('off')
plt.show()
```
